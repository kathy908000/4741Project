## Re-input 

```python
# generate binary matrix

def binary_init(colname, df, dfnew):
    uniques = list(df[colname].unique())
    for u in uniques:
        if type(u) != float:
            this_name = colname + "." + str(u)
            dfnew[this_name] = 0
            print("Generate: ", this_name)
        else:
            if math.isnan(u):
                print("Column ", colname, ' includes NaN.')
            else:
                this_name = colname + "." + str(u)
                dfnew[this_name] = 0
                print("Generate: ", this_name)

def binary_input(colname, df, dfnew):
    uniques = list(df[colname].unique())
    for i in range(len(df)):
        org = df.loc[i, colname]
        if (type(org) == float) and (math.isnan(org)):
            for u in uniques:
                if type(u) == float and math.isnan(u):
                    pass
                else:
                    this_name = colname + "." + str(u)
                    dfnew.loc[i, this_name] = float('nan')
            print("NaN ", i)   
        else:
            change = colname + "." + str(org)
            dfnew.loc[i, change] = 1
            print("Input ", i, ", ", org) 
            

def binary_sing(colname, df, dfnew):
    binary_init(colname, df, dfnew)
    binary_input(colname, df, dfnew)
```

## Input Ordinals 

```python
glft = ha[['Total.Charges', 'Total.Costs', 'Length.of.Stay', 'APR.Severity.of.Illness.Code']]
```

```python
glft
```

```python
# APR.Risk.of.Mortality
hafit['APR.Risk.of.Mortality'].unique()
```

```python
mort = []
for i in range(len(hafit)):
    this_mort = hafit.loc[i, 'APR.Risk.of.Mortality']
    if this_mort == 'Minor':
        mort.append(1)
    elif this_mort == 'Major':
        mort.append(2)
    elif this_mort == 'Extreme':
        mort.append(3)
    elif this_mort == 'Moderate':
        mort.append(4)
glft['APR.Risk.of.Mortality'] = mort
```

```python
# Age.Group
hafit['Age.Group'].unique()
```

```python
age_ord = []
for i in range(len(hafit)):
    this_age = hafit.loc[i, 'Age.Group']
    if this_age == '0 to 17':
        age_ord.append(1)
    elif this_age == '18 to 29':
        age_ord.append(2)
    elif this_age == '30 to 49':
        age_ord.append(3)  
    elif this_age == '50 to 69':
        age_ord.append(4)  
    elif this_age == '70 or Older':
        age_ord.append(5)  
glft['Age.Ordinal'] = age_ord
```

## Input Binary

```python
glft1 = glft.copy()
for colname in colnames[:-2]:
    binary_sing(colname, hafit1, glft1)
```
